#### 1- Encuentra el numero ideal de almacenes 

```python
#tu codigo aquí
from sklearn.cluster import KMeans

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(latlong)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.title('Metodo Elbow usando distorcion')
plt.ylabel("Suma del cuadrado dentro del grupo (WCSS)")
plt.show()
```

Comentario de alumno:


Para determinar el número óptimo de cada grupo, se debe seleccionar el valor de k en el "elbow", es decir, el punto después del cual la distorsión/inercia comienza a disminuir de forma lineal. 
Por lo tanto, podemos concluir que el número óptimo para un buen conjunto o agrupamiento de datos es el numero 3 tal y como lo observamos en la grafica anterior usando el metodo Elbow.


```python
#encontramos las agrupacions
kmeans = KMeans(n_clusters=3)
identified_cluster = kmeans.fit_predict(latlong)

```

```python
# Creamos un nuevo dataframe y agregamos sus "clusters" agrupamientos
df_w_clusters = pd.DataFrame(data=df,columns=['latitude','longitude'])
df_w_clusters['Cluster'] = identified_cluster
df_w_clusters["geometry"] = list(zip(df.longitude, df.latitude))
df_w_clusters["geometry"] = df_w_clusters["geometry"].apply(Point)
df_w_clusters
```

```python
fig, gax = plt.subplots(figsize=(10,10))
world.query("name == 'United States of America'").plot(ax = gax, edgecolor='black', color='white')
plt.scatter(df['longitude'], df['latitude'], c=df_w_clusters['Cluster'], cmap = 'rainbow')
gax.set_xlabel('longitude')
gax.set_ylabel('latitude')
gax.set_title('Target en Estados Unidos')
gax.spines['top'].set_visible(False)
gax.spines['right'].set_visible(False)
plt.show()
```

```python
centers = kmeans.cluster_centers_
df_w_centers = pd.DataFrame(data=centers,columns=['latitude','longitude'])
# df_w_centers['Cluster'] = identified_cluster

# df_w_centers["geometry"] = list(zip(centers.longitude, df.latitude))
# df_w_centers["geometry"] = df_w_centers["geometry"].apply(Point)
df_w_centers
```
